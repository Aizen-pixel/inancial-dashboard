```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Set up the Safari WebDriver
driver = webdriver.Safari()

# Open the webpage
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
driver.get(url)

# Wait for the page to fully load (adjust time if necessary)
time.sleep(5)

# Find all table elements on the page
tables = driver.find_elements(By.TAG_NAME, "table")
print(f"Found {len(tables)} tables on the page.")

# Inspect tables to identify the one containing revenue data
for index, table in enumerate(tables):
    print(f"\nTable {index}:")
    print(table.get_attribute("outerHTML")[:500])  # Print first 500 characters for inspection

# Update the index after identifying the correct table (e.g., tables[1] or tables[2])
target_table_html = tables[1].get_attribute("outerHTML")  # Replace index as needed

# Read the table into a DataFrame
tesla_revenue = pd.read_html(target_table_html)[0]

# Clean the DataFrame
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "").str.replace(",", "")
tesla_revenue.dropna(inplace=True)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(float)

# Display the last five rows
print("Tesla Revenue Data:")
print(tesla_revenue.tail())

# Close the browser
driver.quit()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to create the stock graph
def make_graph(stock_data, title):
    # Plotting the closing price of Tesla stock
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Tesla Stock Price')
    plt.title(title)  # Title for the graph
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Downloading Tesla (TSLA) stock data for the last 1 year (you can adjust the period)
tesla_data = yf.download('TSLA', period='1y')

# Call the make_graph function with the Tesla data and a title
make_graph(tesla_data, 'Tesla Stock Price Over the Last Year')
```
